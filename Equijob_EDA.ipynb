```python
def clean(q):

    words = q.split()

    # Initialize an empty list to store cleaned words
    cleaned_words = []

    # Iterate through the words and check if they are in the list of stop words
    for word in words:
        if word not in stops:
            cleaned_words.append(word)

    # Join the cleaned words into a single string with spaces
    cleaned_text = ' '.join(cleaned_words)

    return cleaned_text
```

```python
from nltk.tokenize import wordpunct_tokenize
def tokenize_text(text):
    return wordpunct_tokenize(text)

# Apply the tokenize_text function to create a new 'Tokenized_Description' column
df['Tokenized_Description'] = df['Job_description'].apply(tokenize_text)
```

```python
df.head()
```

```python
import nltk
from nltk import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
nltk.download('wordnet')
def lemmatize_text(text):
    return " ".join([lemmatizer.lemmatize(word) for word in text])

# Apply the lemmatize_text function to create a new 'Lemmatized_Description' column
df['Lemmatized_Description'] = df['Tokenized_Description'].apply(lemmatize_text)
df['Lemmatized_Description_real'] = df['Lemmatized_Description'].apply(tokenize_text)
```

```python
df.head()
```

```python
def digits_strip(text):
    return re.sub(r'\d', '', text)

# Apply the digits_strip function to create a new column 'tokenized_digits_stripped'
df['tokenized_digits_stripped'] = df['Lemmatized_Description'].apply(digits_strip)
df['tokenized_digits_stripped_real'] = df['tokenized_digits_stripped'].apply(tokenize_text)
```

```python
df['Length_before'] = [len(df['Lemmatized_Description_real'][i]) for i in df.index]
df['Length_after'] = [len(df['tokenized_digits_stripped_real'][i]) for i in df.index]
```

```python
def bias(text):
    male_count = 0
    female_count = 0
    racial_count = 0


    for word in text.split():
        if word in masculine:
            male_count += 1
        if word in feminine:
            female_count += 1
        if word in racial:
            racial_count +=1

    return pd.Series({'male_bias': male_count, 'female_bias': female_count, 'racial_bias' : racial_count})

# Apply the bias function to each row and update 'male_bias' and 'female_bias' columns
df1[['male_bias', 'female_bias', 'racial_bias']] = df['Lemmatized_Description'].apply(bias)
```

```python
df1['% male_bias'] = (df1['male_bias']/df1['Length'])*100
df1['% female_bias'] = (df1['female_bias']/df1['Length'])*100
df1['% racial_bias'] = (df1['racial_bias']/df1['Length'])*100
```
